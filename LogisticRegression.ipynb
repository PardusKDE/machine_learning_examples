```
 import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
csv = pd.read_csv("house.csv")
```

```
csv
```

```
X = np.array(csv["average_price"])
y = np.array(csv["borough_flag"])
```

```
 
plt.style.use("ggplot")
plt.scatter(X, y)
plt.xlabel("Num")
plt.ylabel("Price")
plt.show()
```

```
a, b = np.polyfit(X, y, 1)
```

```
yp = a + b**1 * X**1 + b**2 * X**2
```

```
yp
```

```
pre1 = np.poly1d(yp)
```

```
plt.hist(pre1)
plt.ylabel("Sayi")
plt.xlabel("Sinif")
plt.show()
```
